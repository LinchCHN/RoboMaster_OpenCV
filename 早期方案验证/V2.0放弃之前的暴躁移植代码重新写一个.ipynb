```python
import cv2
import numpy as np
import time
```

```python
# 初始化摄像头
cap = cv2.VideoCapture(4)
```

```python
# 设置摄像头的帧大小
cap.set(cv2.CAP_PROP_FRAME_WIDTH, 1080)
cap.set(cv2.CAP_PROP_FRAME_HEIGHT, 980)
```

```python
lower_blue = np.array([100, 50, 50])
upper_blue = np.array([130, 255, 255])
lower_red = np.array([0, 50, 50])
upper_red = np.array([10, 255, 255])
```

```python
start_time = time.time()
while True:
    # 读取一帧图像
    ret, frame = cap.read()

    if not ret:
        break

    # 转换图像到HSV色彩空间
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # 使用颜色阈值找到蓝色目标
    blue_mask = cv2.inRange(hsv, lower_blue, upper_blue)
    _, blue_contours, _ = cv2.findContours(blue_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # 寻找蓝色目标的区域
    for contour in blue_contours:
        if cv2.contourArea(contour) > 1000:
            # 获取目标的上下两个点
            x, y, w, h = cv2.boundingRect(contour)
            top_point = (x + w // 2, y)
            bottom_point = (x + w // 2, y + h)

            # 在图像中标注出上下两个点
            cv2.circle(frame, top_point, 5, (0, 255, 0), -1)  # 上方点用绿色标注
            cv2.circle(frame, bottom_point, 5, (0, 0, 255), -1)  # 下方点用红色标注

    # 显示结果图像
    cv2.imshow('Frame', frame)

    # 检测按键，如果按下 'q' 键，则退出循环
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break
    
    # 计算帧速率
    end_time = time.time()
    fps = 1 / (end_time - start_time)
    print(f"FPS: {fps}")

    # 更新起始时间
    start_time = time.time()

# 释放摄像头资源
cap.release()

# 关闭所有窗口
cv2.destroyAllWindows()
```
